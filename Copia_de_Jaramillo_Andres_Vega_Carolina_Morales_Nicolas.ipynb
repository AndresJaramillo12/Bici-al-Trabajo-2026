## **PREGUNTA 4**
En días laborales, ¿la media de cnt ≥ 4.500? (plantea y resuelve un test de hipótesis de una muestra).

```python
# 1. Filtrar datos de días laborales
cnt_laborales = df[df['workingday'] == 1]['cnt']
plt.hist(cnt_laborales, bins= 50)
plt.xlabel('cnt')
plt.ylabel('Frecuencia')
plt.title('Distribución de cnt en días laborales')
plt.show()
stats.probplot(df['cnt'], dist="norm", plot=plt)
plt.title("Distribuciòn de bicicletas rentadas")
plt.show()

# Valor de referencia para la prueba
valor_hipotesis = 4500

media_muestral = cnt_laborales.mean()
desviacion_estandar = cnt_laborales.std()
n = len(cnt_laborales)
print (media_muestral, desviacion_estandar, n)


t_statistic, p_value = stats.ttest_1samp(cnt_laborales, valor_hipotesis)
```

**Interpretación pregunta 4**

Al analizar los datos de días laborales, se nota que la distribución de CNT está muy cargada hacia la derecha, como se ve en el histograma y en el gráfico Q-Q. Esto muestra que los datos no son normales, pero como la muestra es grande (más de 11.000 registros), se puede aplicar una prueba paramétrica gracias al Teorema del Límite Central. El promedio real calculado fue cercano a 193, que está muy por debajo del valor de referencia de 4.500 que se planteó en la hipótesis nula. El test de hipótesis dio un p-valor prácticamente igual a cero, lo que nos permite rechazar H₀ y concluir que la media en días laborales no llega a 4.500. En conclusión, el número de bicicletas rentadas entre semana es mucho menor a lo esperado, confirmando la hipótesis alternativa.

## **PREGUNTA 5**
¿Cuál es un IC bootstrap para la mediana de cnt y cómo cambia tu mensaje si la distribución es sesgada?

```python
# Número de remuestreos
n_iterations = 5000
sample_size = len(df['cnt'])
medians = []

for i in range(n_iterations):
    bootstrap_samples = df['cnt'].sample(n=sample_size, replace=True , random_state=rng)
    medians.append(bootstrap_samples.median())

confidence_interval = np.percentile(medians, [2.5, 97.5])

original_median = df['cnt'].median()

print(f"Intervalo de confianza bootstrap para la mediana de 'cnt': {confidence_interval}")
print(f"Mediana de los datos originales: {original_median}")
```

**Interpretación pregunta 5**

El intervalo de confianza bootstrap muestra que la mediana real de bicicletas rentadas está entre 139 y 146, con un valor central de 142. Esto significa que un día típico presenta alrededor de 140 rentas. Dado que la distribución es sesgada, la media no es confiable, pero la mediana refleja mejor la tendencia central. En conclusión, el número de bicicletas usadas en la práctica es muy inferior a 4.500, y los resultados con bootstrap fortalecen esta afirmación al no depender del supuesto de normalidad.
